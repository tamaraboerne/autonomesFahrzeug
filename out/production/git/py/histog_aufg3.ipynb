```python
import matplotlib.pyplot as plt
import numpy as np

# Daten aus Datei lesen
file_path = 'train.txt'
with open(file_path, 'r') as file:
    raw_data = file.read()

# Aufteilen der Daten in Zeilen und Verarbeitung
lines = raw_data.split("\n")
data = {"H": [], "G": []}

for line in lines:
    if line.strip():  # Überspringen leerer Zeilen
        parts = line.split()
        label = parts[0]
        values = list(map(int, parts[1:]))
        data[label].extend(values)

# Erstellen des Histogramms für Zustand H und G
plt.figure(figsize=(10, 6))

if "H" in data:
    plt.hist(data["H"], bins=range(10, 61, 1), alpha=0.5, label="Zustand H", edgecolor='black')
if "G" in data:
    plt.hist(data["G"], bins=range(10, 61, 1), alpha=0.5, label="Zustand G", edgecolor='black')

plt.title("Histogramm für Zustände H und G")
plt.xlabel("Wertebereich")
plt.ylabel("Häufigkeit")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)


# Speichere das Histogramm als Bilddatei
plt.savefig('histogramm_Zustaende.png')
plt.show()
plt.close()
```
